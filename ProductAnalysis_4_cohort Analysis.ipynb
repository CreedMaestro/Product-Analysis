# _01. 데이터 로드

## 1.2. Cohort에 필요한 컬럼만 가져오기

```python
df_base = df_seat[['user_key', 'branch', 'sales_type', 'items', 'paid_date', 'item_seg']]
df_base.head()
```

# _02. data cleansing

## 2.1. 매출만 남기기

```python
df_base['sales_type'].unique().tolist()
```

```python
df_base = df_base[df_base['sales_type'] == '매출']
df_base.info()
```

## 2.2. data type 맞추기

```python
# df_base['paid_date'] = df_base['paid_date'].astype('datetime64[ns]')
df_base['paid_date'] = pd.to_datetime(df_base['paid_date'], format = '%Y-%m-%d')
df_base.info()
```

```python
df_base.head(2)
```

# _03. 분석하기 좋은 data로 만들기

## 3.1. 최초 구매일 컬럼 생성

```python
df_base.head(10)
```

```python
df_base['first_paid'] = df_base.groupby(['user_key'])['paid_date'].transform('min')
df_base.head()
```

## 3.2. 월 단위 형태로 data 맞추기

```python
df_base['first_ym'] = df_base['first_paid'].dt.strftime('%Y-%m') # yyyy-mm
df_base['paid_ym'] = df_base['paid_date'].dt.strftime('%Y-%m')
df_base.head()
```

### 3.2.1 data range 확인

```python
df_base['paid_ym'].nunique()  ## paid_ym컬럼 '21년 자료'. 1년이면 12달인데 13개의 값이 나왔다? 재확인 필요
```

```python
print(df_base['paid_ym'].min())
print(df_base['paid_ym'].max())
```

```python
df_base.groupby(['paid_ym']).agg(count = ('user_key', 'nunique')).reset_index()  #2022-21 22개 행 확인
```

```python
# 12월 31일에서 1월 1일 넘어가는 자정에 구매시 2022-01로 잡혔다고 생각. 22건은 12월로 합산
df_base['paid_ym'] = df_base['paid_ym'].replace('2022-01', '2021-12')
df_base['first_ym'] = df_base['first_ym'].replace('2022-01', '2021-12')
```

```python
print(df_base['paid_ym'].min())
print(df_base['paid_ym'].max())
```

```python
df_base.head()
```

## 3.3. 매 월 단위 동질 집단 리텐션 숫자 생성

```python
df_user_cnts = df_base.groupby(['first_ym', 'paid_ym'])['user_key'].agg('nunique').reset_index()
df_user_cnts = df_user_cnts.rename(columns = {'user_key':'user_cnts'})
df_user_cnts
```

### 3.3.1. 매 월 데이터가 있는 지 확인 해 보기

Screen Shot 2022-05-26 at 14.48.58.png

* 현재 데이터는 1년 12달에 대한 cohort를 보기 위한 것 이므로,
    * 12 + 11 + 10 + ... + 3 + 2 + 1 = 총 row 개수 78과 맞는 지 확인 해 보면 된다
    * ( 12 + 1 ) + ( 11 + 2 ) + ... + ( 7 + 6 ) = 13 x 6 = 78

```python
# 실제로 월마다 모든 cohort 개수가 있는 지 확인
df_user_cnts['first_ym'].value_counts()
```

### 3.3.2. Cohort 기간 컬럼 생성

```python
df_user_cnts.head(10)
```

```python
# 실제로 월마다 모든 cohort 개수가 있는 지 확인되었으므로, Cohort 기간 컬럼 생성
df_user_cnts['cohort_period'] = df_user_cnts.groupby(['first_ym']).cumcount()+1
df_user_cnts.head()
```

```python
# user_cnts: numeric data type 여부 확인
df_user_cnts.info()
```

### 3.3.3. 기준이 될 것을 index로, 값으로 표현될 것을 컬럼으로 만들기

```python
df_user_cnts
```

```python
df_user_cnts.set_index(['first_ym', 'cohort_period'])
```

### 3.3.4. 값을 기준으로 unstack

```python
user_cohort = df_user_cnts.set_index(['first_ym', 'cohort_period'])['user_cnts'].unstack()#.fillna(' ')
#user_cohort = df_user_cnts.set_index(['first_ym', 'paid_ym'])['user_cnts'].unstack()
user_cohort
```

```python
df_user_cnts.set_index(['first_ym', 'cohort_period'])['user_cnts']
```

### 3.3.5. unstack 결과를, 첫 코호트 기간의 값으로 나눠주기 → retention 산출

```python
user_cohort[0]  #첫번째 컬럼 확인
```

```python
user_retention = user_cohort.divide(user_cohort[0], axis = 0)
user_retention
```

# _04. 시각화

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (18, 6))

sns.heatmap(
    data = user_retention,
    annot = True,
    fmt = '.0%', #소수점 자릿수
    linecolor = 'white',
    linewidth = 0.1,
    # cmap = 'Spectral'
    cmap = 'Greens'
)
plt.title(f'\nRetention Cohort Visualization\n', fontsize = 20, loc = 'left')
plt.show()
```
